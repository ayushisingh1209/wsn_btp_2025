```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Load dataset
df = pd.read_csv(r"C:\Users\VIVEK KUMAR SINGH\Downloads\reshaped_data.csv")

# Replace blank (empty) values with 0
df.replace(r'^\s*$', 0, regex=True, inplace=True)

# Replace NaN values with 0
df.fillna(0, inplace=True)
print(df)

```

```python
# Prepare features (X) and target (y)
X, y = [], []

for i in range(len(df)):
    node_data = df.iloc[i, 1:].values  # Ignore 'id'
    for t in range(len(node_data) - 1):
        X.append(node_data[:t + 1])  # Past values
        y.append(node_data[t + 1])   # Next value
```

```python
# Pad sequences to the same length
max_length = max(len(seq) for seq in X)
X_padded = np.array([np.pad(seq, (0, max_length - len(seq)), 'constant') for seq in X])
```

```python
# Convert y to a NumPy array
y = np.array(y)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)
```

```python
# Train the model
# model = RandomForestRegressor(n_estimators=100, random_state=42)
model = RandomForestRegressor(n_estimators=50, max_depth=10, max_features='sqrt', random_state=42, n_jobs=-1)
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate model performance
mae = mean_absolute_error(y_test, y_pred)
print(f"\nModel Performance: MAE = {mae:.4f}")
```

```python
# Predict next energy consumption per node
predictions, next_timestamps = [], []

for i in range(len(df)):
    node_data = df.iloc[i, 1:].values
    # Ensure that node_data has the correct length (apply padding if necessary)
    if len(node_data) < 195:  # Ensure the data has exactly 195 features
        node_data_padded = np.pad(node_data, (0, 195 - len(node_data)), 'constant')
    elif len(node_data) > 195:
        node_data_padded = node_data[:195]  # Trim data to ensure it matches 195 features
    else:
        node_data_padded = node_data  # No padding needed if node_data is already equal to 195

    # Reshape the data to match the training input shape
    node_data_padded = node_data_padded.reshape(1, -1)
    
    # Ensure that the number of features matches
    if node_data_padded.shape[1] != 195:
        print(f"Feature mismatch for node {df.iloc[i, 0]}: Expected 195 features, but got {node_data_padded.shape[1]}")
        continue
    
    # Predict the next energy consumption
    prediction = model.predict(node_data_padded)
    predictions.append(prediction[0])

    # Get next timestamp
    last_time = timestamps[len(node_data) - 1]
    next_time_index = timestamps.index(last_time) + 1 if last_time in timestamps else None
    next_time = timestamps[next_time_index] if next_time_index and next_time_index < len(timestamps) else "Unknown"
    next_timestamps.append(next_time)

# Check the lengths of predictions and next_timestamps
print(f"Length of predictions: {len(predictions)}")
print(f"Length of next_timestamps: {len(next_timestamps)}")
print(f"Length of df: {len(df)}")

# Save predictions if lists are not empty
if len(predictions) > 0 and len(next_timestamps) > 0:
    output_df = pd.DataFrame({'Node_ID': df.iloc[:, 0], 'Next_Timestamp': next_timestamps, 'Predicted_Next_Energy': predictions})
    output_df.to_csv(r"C:\Users\VIVEK KUMAR SINGH\Downloads\predicted_energy_with_timestamp.csv", index=False)
    print("\nPredictions saved successfully!")
else:
    print("\nNo predictions or timestamps generated. Please check the loop logic.")
```
